## 1. What is Semantic Kernel?

**Semantic Kernel (SK)** is Microsoft's **lightweight SDK** for integrating LLMs into conventional applications. It acts as a **bridge between AI models and enterprise code** вҖ” letting you combine AI calls with regular functions, business logic, and existing systems.

Its enterprise edge is **mixed-language support** (strong **.NET/C#**, plus Python and Java) and tight integration with **Azure OpenAI**. In 2026 it's one half of the new **Microsoft Agent Framework** (SK's stability + AutoGen's orchestration).

> Think of SK as the **orchestration glue**: it decides *which* AI/code capabilities to call and chains them into sophisticated workflows.

## 3. The Agent Framework

The **Semantic Kernel Agent Framework** lets you build AI agents inside the SK ecosystem. An SK agent is a software entity that:

- Receives input, processes information, and takes actions toward a goal
- Can **send and receive messages** (multi-agent conversations)
- Generates responses using models, tools (plugins), and customizable components

2026 improvements include **more reliable function calling** (structured parameters) and a **smarter planner** that decomposes tasks and selects execution strategies more efficiently вҖ” plus the roadmap toward full multi-agent systems under the Microsoft Agent Framework.

```python
import semantic_kernel as sk
from semantic_kernel.connectors.ai.open_ai import AzureChatCompletion

# 1. Create the kernel and attach a model connector
kernel = sk.Kernel()
kernel.add_service(AzureChatCompletion(deployment_name="gpt-4o", ...))

# 2. Define a plugin (a reusable skill the AI can call)
class WeatherPlugin:
    @sk.kernel_function(description="Get the weather for a city")
    def get_weather(self, city: str) -> str:
        return weather_api.lookup(city)

kernel.add_plugin(WeatherPlugin(), plugin_name="weather")

# 3. The planner decomposes a goal and calls plugins automatically
result = await kernel.invoke_prompt(
    "Plan a weekend trip to a city with good weather this weekend."
)
print(result)
```

> The **planner** turns a high-level goal into a sequence of plugin calls вҖ” you provide capabilities, SK figures out the orchestration.

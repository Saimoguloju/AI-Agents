## 1. What is Mirascope?

**Mirascope** calls itself **"The LLM Anti-Framework."** It's a Python library offering a **"Goldilocks API"** — the **fine-grained control** of raw provider APIs *plus* the **type safety and ergonomics** of higher-level frameworks, with nothing heavier in between.

Its analogy: if provider-native APIs are **HTML/CSS** and big agent frameworks are **Angular**, Mirascope aims to be **React** — light, composable, and close to the metal.

> It resists lock-in: you define calls, tools, prompts, and structured outputs in plain Python, and stay in control of the flow.


<a id="2"></a>
## 2. The "Anti-Framework" Philosophy

```
 raw provider APIs ◀───────────── Mirascope ─────────────▶ heavy frameworks
   (max control,                 (Goldilocks:               (max abstraction,
    no ergonomics)            control + ergonomics)          less control)
```

The pitch: most frameworks make you adopt their abstractions and "magic." Mirascope keeps you **writing normal Python** — it just removes the boilerplate of prompts, structured parsing, retries, and provider differences, **without hiding the control flow** from you.

## 4. Conceptual Code

```python
from mirascope import llm
from pydantic import BaseModel

# 1. Strict structured output — returns a typed Pydantic object
class Book(BaseModel):
    title: str
    author: str

# 2. A plain decorated function IS the LLM call (close to the metal)
@llm.call(provider="anthropic", model="claude-opus-4-8", response_model=Book)
def extract_book(text: str) -> str:
    return f"Extract the book from: {text}"

book = extract_book("I loved Dune by Frank Herbert.")
print(book.title, "—", book.author)   # validated, type-hinted

# Swap provider="openai", model="gpt-4o" with no structural rewrite.
```

> No graph, no orchestration engine — just typed Python functions that wrap LLM calls, with structured outputs and provider-switching handled for you.

## 5. How Mirascope is Advanced vs Other Frameworks

| Dimension | Mirascope | Pydantic AI | LangGraph | CrewAI |
|---|---|---|---|---|
| **Philosophy** | **Anti-framework (Goldilocks)** | Type-safe agents | Graph control | Role teams |
| **Abstraction** | **Minimal, close to APIs** | Moderate | Higher | High |
| **Structured output** | **Pydantic / JSON mode** | Pydantic (core) | DIY | Basic |
| **Control flow** | **You keep it (explicit)** | Agent-managed | Graph | Process engine |
| **Multi-provider** | **Yes, no rewrite** | Yes | Adapters | Adapters |

### What makes it "advanced"
1. **Goldilocks positioning** — uniquely targets developers who want **raw-API control without raw-API boilerplate**, a real gap between "just use the API" and "adopt a big framework."
2. **Type-safe structured output** with Pydantic, like Pydantic AI, but with an even **lighter, less opinionated** footprint.
3. **No hidden control flow** — you always own the orchestration, which makes behavior easy to reason about and debug.
4. **Provider-agnostic + observability** — switch models freely and get versioning, tracing, and cost tracking from a hosted dashboard.

## 6. Strengths, Weaknesses & When to Use

### ✅ Strengths
- Maximum **control** with minimal boilerplate
- **Type-safe structured outputs** (Pydantic)
- No lock-in; switch providers without rewrites
- Built-in tracing / cost tracking

### ⚠️ Weaknesses
- **You build orchestration yourself** (not batteries-included for multi-agent)
- Less hand-holding than CrewAI/LangGraph
- Smaller ecosystem/community

### 🎯 Use Mirascope when…
- You want to stay **close to the model API** but skip boilerplate
- **Structured, validated outputs** matter and you dislike heavy frameworks
- You value **explicit control** over magic orchestration

> **Rule of thumb:** *If big frameworks feel like too much magic but the raw API feels like too much boilerplate, Mirascope is the middle path.*

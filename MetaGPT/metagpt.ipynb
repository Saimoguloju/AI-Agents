## 1. What is MetaGPT?

**MetaGPT** is a multi-agent framework that **simulates an entire software company**. Give it a **one-line requirement**, and it produces user stories, competitive analysis, requirements, data structures, APIs, documents — and code — by internally running **product managers, architects, project managers, engineers, and QA**.

Its insight: software quality comes from **well-defined Standard Operating Procedures (SOPs)** executed by specialized roles — not from a single all-knowing agent.

> MetaGPT turns the messy "one agent does everything" approach into a **structured assembly line** of role-specialized agents.


<a id="2"></a>
## 2. The Core Thesis: Code = SOP(Team)

```
        Code  =  SOP( Team )
                  │      │
        encoded procedures   specialized roles
```

- **SOPs are encoded into prompt sequences**, giving each step **execution focus**.
- **Specialized roles** enhance relevance (each agent does one job well).
- **Standardized outputs** create clear dependencies between agents.
- This structure **minimizes the "cascading hallucination problem"** — where one agent's error snowballs through a naive multi-agent chain.

> Rather than hoping agents self-organize, MetaGPT **bakes in the workflow of a real software team**.


<a id="3"></a>
## 3. The Software Company Roles

MetaGPT defines **five roles**, each inheriting from a base `Role` class that encodes both LLM interaction *and* job-specific behavior:

```
   ProductManager ─▶ Architect ─▶ ProjectManager ─▶ Engineer ─▶ QA Engineer
      analyzes        designs        plans &           writes       tests &
      requirements    architecture   breaks down       code         verifies
```

| Role | Responsibility |
|---|---|
| **Product Manager** | Requirement & feasibility analysis |
| **Architect** | System blueprint / software architecture |
| **Project Manager** | Task breakdown & coordination |
| **Engineer** | Writes the actual code |
| **QA Engineer** | Tests and verifies output |


<a id="4"></a>
## 4. How It Works

Agents communicate through an **assembly-line paradigm** with a **publish-subscribe** pattern — each agent produces **structured outputs** that become the input prompt for the next:

```
Human request
     │
     ▼
ProductManager ──(PRD)──▶ Architect ──(design)──▶ ProjectManager
                                                       │ (task list)
                                                       ▼
                                 Engineer ──(code)──▶ QA Engineer ──▶ Final app
```

When MetaGPT receives a request, the **ProductManager** starts with requirement + feasibility analysis; the **Architect** produces a detailed system blueprint; downstream roles implement and verify. The shared message pool (publish-subscribe) keeps everyone coordinated.

## 7. Strengths, Weaknesses & When to Use

### ✅ Strengths
- Purpose-built for **end-to-end software generation**
- SOP + structured outputs improve reliability & coordination
- Clear, human-like role division (PM → Architect → … → QA)
- Strong research foundation

### ⚠️ Weaknesses
- **Narrow domain** — optimized for software projects, not general agents
- Less flexible than general orchestrators for arbitrary tasks
- Output still needs human review for complex/production code

### 🎯 Use MetaGPT when…
- You want to **prototype software from a one-line spec**
- The task maps to a **software-team SOP** (requirements → design → code → QA)
- You're researching **structured multi-agent collaboration**

> **Rule of thumb:** *Reach for MetaGPT when the job literally is "build me software" and you want a team-with-a-process, not a single coder agent.*

## 1. What is Mastra?

**Mastra** is a **TypeScript-native** framework for building AI agents and applications, from the team behind **Gatsby**. It's the **full production toolkit**: agents that reason and use tools, deterministic workflows, RAG, persistent memory, evals, and observability — all in TypeScript.

It's built **on top of the Vercel AI SDK** (which handles low-level model calls and streaming) and adds the **higher-level abstractions production apps need**.

> Crucially, Mastra is **TypeScript-native from scratch** — not a port of a Python framework — so it feels idiomatic to JS/TS teams.


<a id="2"></a>
## 2. What It Adds on Top of the AI SDK

```
   ┌──────────────────────────────────────────┐
   │                 MASTRA                    │  ← agents, workflows, RAG,
   │   (production abstractions, TS-native)    │     memory, evals, MCP
   ├──────────────────────────────────────────┤
   │            Vercel AI SDK                  │  ← model calls + streaming
   ├──────────────────────────────────────────┤
   │   OpenAI · Anthropic · Google · 90+ ...   │  ← 3,300+ models, 94 providers
   └──────────────────────────────────────────┘
```

The Vercel AI SDK handles the **model layer**; Mastra adds **agents, orchestration, knowledge, memory, and ops** — the parts you'd otherwise build yourself.

## 4. Conceptual Code

```typescript
import { Mastra } from "@mastra/core";
import { Agent } from "@mastra/core/agent";
import { openai } from "@ai-sdk/openai";

// 1. Define an agent (tools + instructions + memory)
const supportAgent = new Agent({
  name: "support",
  instructions: "Help customers; use tools to look up orders.",
  model: openai("gpt-4o"),       // type-safe model router
  tools: { getOrderStatus },     // typed tools
});

// 2. Register with Mastra (adds memory, workflows, observability)
export const mastra = new Mastra({ agents: { supportAgent } });

// 3. Run it
const result = await supportAgent.generate("Where is order #4821?");
console.log(result.text);
```

> You get the Vercel AI SDK's model layer **plus** Mastra's memory, workflows, evals, and one-command deploy — all type-safe.

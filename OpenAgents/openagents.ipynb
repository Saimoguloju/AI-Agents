## 1. What is OpenAgents?

**OpenAgents** builds **persistent agent networks** — self-sustaining communities where agents **discover peers, collaborate, and grow over time**. You create a network, add agents, and they **autonomously discover each other's capabilities** and work together. Networks are persistent: agents can **join or leave at any time**.

Its standout claim in 2026: it's the **only framework with native support for both MCP and A2A** — the two open protocols defining how agents talk to *tools* and to *each other*.

> Where most frameworks focus on a single app's agents, OpenAgents targets the emerging **"agent economy"** — many independent agents interoperating across a network.

## 2. MCP vs A2A: Vertical and Horizontal

The two protocols solve **different axes** of interoperability:

```
                 ┌─────────────┐
        A2A ────▶│   Agent A   │◀──── A2A   (horizontal:
       (agent-   └──────┬──────┘             agent ↔ agent)
        to-agent)       │ MCP (vertical: agent ↔ tools/data)
                        ▼
              ┌───────────────────┐
              │ Tools · Data · APIs│
              └───────────────────┘
```

| Protocol | Axis | Standardizes | Origin |
|---|---|---|---|
| **MCP** (Model Context Protocol) | **Vertical** | How an agent connects to **tools & data** | Anthropic → Linux Foundation |
| **A2A** (Agent2Agent) | **Horizontal** | How agents **discover & talk to each other** | Google + 50+ partners |

Supporting **both** means an OpenAgents network can use any MCP tool *and* coordinate with any A2A-speaking agent.

## 3. Persistent Agent Networks

```
        ┌───────────────── OpenAgents Network ─────────────────┐
        │   (persistent — agents join/leave anytime)           │
        │                                                      │
        │   Agent A ◀──A2A──▶ Agent B ◀──A2A──▶ Agent C         │
        │     │ MCP             │ MCP             │ MCP          │
        │     ▼                 ▼                 ▼             │
        │   tools             tools             tools           │
        └──────────────────────────────────────────────────────┘
```

- **Discovery** — agents advertise capabilities; peers find them via A2A.
- **Collaboration** — agents delegate and coordinate autonomously.
- **Persistence** — the network outlives any single task or agent.
- **Contextual tool use** — MCP gives each agent scalable, standardized access to tools and data.

## 4. Conceptual Code

```python
# Conceptual — OpenAgents centers on networks + protocol interop
from openagents import Network, Agent

# 1. Create a persistent network
network = Network(name="research-net")

# 2. Add agents; each exposes capabilities via A2A and uses tools via MCP
researcher = Agent(name="researcher", mcp_servers=["web-search", "arxiv"])
analyst    = Agent(name="analyst",    mcp_servers=["python-sandbox"])

network.add(researcher)
network.add(analyst)

# 3. Agents discover each other (A2A) and collaborate autonomously
network.run(goal="Survey 2026 agent frameworks and produce a comparison.")
# researcher gathers sources (MCP tools) → hands off to analyst (A2A) → result
```

> The emphasis isn't a single hard-coded pipeline — it's **interoperable agents** that find and use each other through open protocols.

## 6. Strengths, Weaknesses & When to Use

### ✅ Strengths
- Native **MCP + A2A** → maximum interoperability
- **Persistent, discoverable** agent networks
- Cross-framework / cross-vendor collaboration
- Aligned with where open standards are heading

### ⚠️ Weaknesses
- Younger, smaller ecosystem than LangChain
- Network/protocol model is **newer territory** (fewer production case studies)
- Overkill for a simple single-app agent

### 🎯 Use OpenAgents when…
- You need **agents from different systems to interoperate**
- You're building a **persistent network** of collaborating agents
- **MCP + A2A** standardization matters to your architecture

> **Rule of thumb:** *If your future involves many independent agents discovering and working with each other, OpenAgents is built for that world.*

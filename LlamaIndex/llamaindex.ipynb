## 1. What is LlamaIndex?

**LlamaIndex** is the leading framework for **connecting LLMs to your data** — and in 2026 it's far more than RAG: it's an **agentic document-processing** platform. It shines whenever **data access sits at the core** of your agent's job.

It provides sophisticated retrieval that goes well beyond simple vector search, plus **AgentWorkflows** for multi-step orchestration and **300+ integrations** (LLMs, embeddings, vector stores).

> If LangGraph is about *control flow* and CrewAI is about *teams*, LlamaIndex is about **grounding agents in knowledge** — documents, databases, and tools.

## 2. From Naive RAG to Agentic RAG

Traditional RAG is **reactive**: retrieve once, then answer. **Agentic RAG** is **proactive**:

```
Naive RAG:     Query ──▶ Retrieve ──▶ Answer

Agentic RAG:   Query ──▶ Plan ──▶ Tool Use ──▶ Reflect ──▶ Answer
                            ▲                       │
                            └──── iterate ──────────┘
```

Here the LLM is a **router, planner, and reasoner**: it can rewrite a bad query, choose which index/tool to hit, retrieve, **reflect** on whether the answer is good, and loop.

> 2026 shift: with good **filesystem tools**, agents do *dynamic search* over document collections — interleaving search and read like a human browsing files — which often beats naive semantic chunking.

## 3. AgentWorkflow & the RAG Pipeline

LlamaIndex separates the pipeline into **distinct, controllable layers**:

```
┌────────────┐   ┌──────────┐   ┌───────────┐   ┌──────────────┐
│ Ingestion  │──▶│ Indexing │──▶│ Retrieval │──▶│ Query Engine │
│ load docs  │   │ vectors/ │   │ + rerank  │   │ synthesize   │
│            │   │ indexes  │   │           │   │ grounded ans │
└────────────┘   └──────────┘   └───────────┘   └──────────────┘
        used as TOOLS by ─────────────▶  AgentWorkflow
```

- **Query Engine** supports response synthesis across nodes, **reranking**, and **iterative query refinement**.
- **AgentWorkflow** orchestrates multi-step agents that call these query engines as tools — enabling end-to-end document automation.

## 4. Conceptual Code

```python
from llama_index.core import VectorStoreIndex, SimpleDirectoryReader
from llama_index.core.agent.workflow import FunctionAgent
from llama_index.core.tools import QueryEngineTool

# 1. Ingest + index your documents
docs = SimpleDirectoryReader("./company_docs").load_data()
index = VectorStoreIndex.from_documents(docs)

# 2. Turn the index into a TOOL the agent can query
query_tool = QueryEngineTool.from_defaults(
    query_engine=index.as_query_engine(),
    name="company_knowledge",
    description="Answer questions about internal company documents.",
)

# 3. An agent that decides WHEN to retrieve (agentic RAG)
agent = FunctionAgent(
    tools=[query_tool],
    llm=llm,
    system_prompt="Use company_knowledge to answer grounded questions.",
)

response = await agent.run("What is our 2026 refund policy?")
```

> The agent doesn't blindly retrieve — it **reasons about when** to call the knowledge tool, the essence of agentic RAG.

## 6. Strengths, Weaknesses & When to Use

### ✅ Strengths
- Best-in-class **RAG / retrieval / document processing**
- Agentic RAG: plan → retrieve → reflect → answer
- Huge connector ecosystem (300+)
- Plays well as a "knowledge tool" inside other frameworks

### ⚠️ Weaknesses
- **Not** primarily a general multi-agent orchestrator
- Overkill if your agent doesn't need heavy data access
- Best results require thoughtful data/index design

### 🎯 Use LlamaIndex when…
- Your agent's value is **grounded answers over your data**
- You need advanced retrieval (rerank, multi-source, document OCR)
- You want to add a strong **knowledge tool** to a LangGraph/CrewAI agent

> **Rule of thumb:** *If "talk to my documents/data" is the hard part, LlamaIndex is the right tool — often alongside another orchestrator.*

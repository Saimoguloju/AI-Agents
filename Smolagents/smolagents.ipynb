## 1. What is smolagents?

**smolagents** is Hugging Face's **minimalist** agent library: build and run capable agents in just a few lines of code. Its philosophy is radical simplicity — **the core agent logic is ~1,000 lines**, with abstractions kept to the thinnest possible layer above raw code.

Its signature move: agents that **think and act by writing Python code**, not by emitting rigid JSON tool calls.

> "Small" is the point. Where other frameworks add orchestration layers, smolagents trusts the LLM to **write code** that composes tools naturally.


<a id="2"></a>
## 2. The Big Idea: CodeAgent

The default agent type is the **CodeAgent**. Instead of producing a structured tool call, it **writes a Python snippet** that calls tools, loops, branches, and computes — then executes it.

```
Traditional tool-calling agent:        CodeAgent:
  step 1: call tool A  (LLM call)         write ONE code block:
  step 2: call tool B  (LLM call)            results = []
  step 3: call tool C  (LLM call)            for x in tool_A():
  ...one step per round-trip                     results.append(tool_B(x))
                                                 answer = summarize(results)
                                             → executed in a sandbox
```

The agent lays out a **whole plan as code at once**, executed efficiently — rather than revealing one tiny step per LLM round-trip.


<a id="3"></a>
## 3. Why "Acting in Code" Wins

Code is a more **natural, composable** action space for an LLM than JSON:

| Property | Benefit |
|---|---|
| **Expressive** | Native loops, conditionals, function nesting — complex logic in one shot |
| **Efficient** | Consolidates many tool calls into one snippet → ~**30% fewer steps & LLM calls** |
| **Accurate** | Superior performance on complex benchmarks vs JSON tool-calling |
| **Flexible** | No need to predefine every action; the agent generates new logic dynamically |
| **Emergent reasoning** | Ideal for multi-step, dynamic problems |

**Security:** generated code runs in a **sandbox** — via E2B, Modal, Blaxel, or Docker — so execution stays isolated.

## 4. Conceptual Code

```python
from smolagents import CodeAgent, WebSearchTool, InferenceClientModel

# 1. Pick a model (any HF/Inference or API model) and some tools
model = InferenceClientModel()          # e.g. a Hugging Face model
agent = CodeAgent(
    tools=[WebSearchTool()],
    model=model,
    # generated code runs sandboxed (e2b / docker / modal)
)

# 2. Run — the agent WRITES PYTHON that uses the tools, then executes it
agent.run(
    "How many seconds would a leopard at full speed take to run the length "
    "of the Golden Gate Bridge?"
)
# Internally it may write code that searches for speed + bridge length,
# then computes the answer — all in one executable snippet.
```

> The agent's "action" is literally a Python program — searching, computing, and combining tools in a single block.

## 6. Strengths, Weaknesses & When to Use

### ✅ Strengths
- Tiny, readable, hackable core (~1,000 LOC)
- Code-action agents: efficient + accurate on complex tasks
- Easy access to **open / Hugging Face models**
- Sandboxed code execution out of the box

### ⚠️ Weaknesses
- Less of a full **multi-agent orchestration** suite
- Code-execution model needs a sandbox (security setup)
- Fewer enterprise/observability features than LangGraph

### 🎯 Use smolagents when…
- Tasks are **computational / tool-heavy / research-y**
- You want a **minimal, transparent** framework you can fully understand
- You're in the **Hugging Face / open-model** ecosystem

> **Rule of thumb:** *If you want the smallest possible agent that "thinks in code," smolagents is the cleanest option.*
